###### Missing value treatment

After replacing the special characters and irregular data, time to impute the missing values. If we check the missing values in every variable/columns, almost every column has some missing values  

```python
df.info()
```

Here is the list of all columns and the number of missing values.

```python
df.isnull().sum()
```

As we have observed that dataset has a lot of numerical columns even if they are categorical in nature in general. We are going to use this as the condidtion in the for loop and using KNN imputer to impute the missing values. 

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
for i in df.columns:
    if df[i].dtype != 'object':
        df[i] = imputer.fit_transform(df[[i]])
        df[i]=df[i].round()
        print('missing value in ' + i + (' imputed with KNNImputer'))
```

Here AccountId nither have any missing values nor a relevent column for data cleaning or EDA. But it gets into the for loop just because of its data type. WE NEED TO IGNORE IT

To impute the values in those columns which the categorical in nature and Object as the datatype, Mode of the column is used. 

```python
for i in df.columns:
    if df[i].dtype == 'object':
        x=df[i].mode()[0]
        df[i]=df[i].fillna(x)
        print('missing value in ' + i + (' with mode of ' + i))
        print(x)
```

Finally we have left with no missing values in the dataset.

```python
def treat_outliers(column):
    # Calculate the first quartile (Q1)
    Q1 = column.quantile(0.25)
    # Calculate the third quartile (Q3)
    Q3 = column.quantile(0.75)
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Treat outliers by capping them to the lower and upper bounds
    treated_col = column.clip(lower=lower_bound, upper=upper_bound)
    return treated_col
```

```python
for i in df_cont_col:
    df[i]= treat_outliers(df[i])
```

```python
def model_evaluation(a,b,c):
    v_name = [name for name, value in locals().items() if name != 'self']
    print('Model score')
    print(c.score(a,b))
    # predict probabilities
    probs = c.predict_proba(a)
    # keep probabilities for the positive outcome only
    probs = probs[:, 1]
    # calculate AUC
    auc = roc_auc_score(b, probs)
    print('Area under the curve')
    print('AUC: %.3f' % auc)
    print('ROC curve')
    # calculate roc curve
    plt.title(' ROC Curve')
    fpr, tpr, thresholds = roc_curve(b, probs)
    plt.plot([0, 1], [0, 1], linestyle='--')
    # plot the roc curve for the model
    plt.plot(fpr, tpr)
    plt.show()
    print('Confusion Matrix')
    d=c.predict(a)
    cm = confusion_matrix(b, d)
    # Assuming 'model.classes_' contains the class labels
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=c.classes_)
    # Plotting the confusion matrix
    disp.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix' )
    plt.show()
    print('Classification Report')
    print(classification_report(b,d))
    TN, FP, FN, TP = cm.ravel()
    # Calculate FPR and TPR
    FPR = FP / (FP + TN)
    TPR = TP / (TP + FN)
    print('False positive rate = ', FPR)
    print('True positive rate = ', TPR)
```

Here a function is defined and we need to provide the argument like "X variables" as 'a',"Y Variable" as 'b', and The "Model" as 'c'. This model_evaluation will provide the evaluation matrices like Score, AUC, ROC Curve, confusion matrix, and classification report as output.

###### Model evaluation for logistic regression on train data

```python
model_evaluation(x_train,y_train, model_logit)
```

###### Model evaluation for logistic regression on test data

```python
model_evaluation(x_test,y_test,model_logit)
```

The model perofrmance is same for the train data and test data. Here the class we are intrested in is 1 that is the churned user. And for this class the performance is not up to the mark.
- For class 1, precision is 0.76, indicating that 76% of the instances predicted as class 1 are actually class 1
- For class 1, recall is 0.47, indicating that only 47% of the actual class 1 instances are correctly predicted as class 1.

#### 1.2 Model Tuning - Logistic regression

The precision and recall for class 0 is very high but the performance of the model is really bad for class 1. This my be because of the imbalance in data. 

Here we are going to use SMOTE to balance the data.

```python
from imblearn.over_sampling import SMOTE 
```
